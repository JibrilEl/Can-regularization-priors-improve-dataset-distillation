## Introduction

This code is to plot in a nicer way the results obtained in the other notebooks

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
accuracy_random_sampling = [0.501, 0.800, 0.893, 0.948, 0.973]
n_data_points = [10, 200, 1000, 10000, 60000]

accuracy_distillation = [0.844, 0.856, 0.81]
n_data_points_distillation = [10, 200, 1000]

accuracy_distillation_smoothness = [0.833, 0.84]
n_data_points_distillation_smoothness = [10, 200]

accuracy_distillation_sparsity = [0.835, 0.837]
n_data_points_distillation_sparsity = [10, 200]

accuracy_distillation_pretrained = [0.836]
n_data_points_distillation_pretrained = [10]

plt.semilogx(n_data_points, accuracy_random_sampling, label = "Random sampling" )
plt.semilogx(n_data_points_distillation, accuracy_distillation, c= 'r', label = "Dataset distillation" )
plt.semilogx(n_data_points_distillation_smoothness, accuracy_distillation_smoothness, c= 'black', label = "Dataset distillation with smoothness prior" )
plt.semilogx(n_data_points_distillation_sparsity, accuracy_distillation_sparsity, c= 'darkgreen', label = "Dataset distillation with sparsity prior" )

plt.semilogx(n_data_points_distillation_pretrained, accuracy_distillation_pretrained, "x",c= 'purple', label = "Dataset distillation with pretrained model prior" )
plt.grid()
plt.xlabel("Number of images")
plt.ylabel("Test set accuracy")
plt.legend()
plt.show()
```
